## Unpacking the 'insurance.csv' file

```python
# open the file, store it into a list, then enumerate the list and print first five elements
with open("insurance.csv") as insurance_file:
    insurance_file_list = list(csv.DictReader(insurance_file))
    for i, row in enumerate(insurance_file_list):
        if i < 5:
            print(i, row)
        else:
            break

# length of the file is equal to number of medical insurance users
print(len(insurance_file_list))
```

## Organizing the data into lists for easier manipulation

```python
# an empty list corresponding to every column name
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

# open the file, store it into a dictionary, for every column name, store the corresponding row into a list
with open("insurance.csv") as insurance_file:
    insurance_file_dictionary = csv.DictReader(insurance_file)
    for row in insurance_file_dictionary:
        age.append(row["age"])
        sex.append(row["sex"])
        bmi.append(row["bmi"])
        children.append(row["children"])
        smoker.append(row["smoker"])
        region.append(row["region"])
        charges.append(row["charges"])

# check how the lists look in an arbitrarily chosen range        
print(age[365:375])
print(sex[365:375])
print(bmi[365:375])
print(children[365:375])
print(smoker[365:375])
print(region[365:375])
print(charges[365:375])
```

Every list containing numbers has them saved as strings.

## Transform lists

```python
# convert list containing strings as integers to a list of integers
age_clean = []
for a in age:
    age_clean.append(int(a))
    
children_clean = []
for c in children:
    children_clean.append(int(c))

# convert list containing strings as floats to a list of floats
bmi_clean = []
for b in bmi:
    float_b = float(b.replace("'", ""))
    bmi_clean.append(float_b)
    
charges_clean = []
for c in charges:
    float_c = float(c.replace("'", ""))
    charges_clean.append(float_c)

# sort the list from smallest number to highest
age_sorted = sorted(age_clean)

# check how the lists look in an arbitrarily chosen range  
print(age_clean[500:510])
print(sex[500:510])
print(bmi_clean[500:510])
print(children_clean[500:510])
print(smoker[500:510])
print(region[500:510])
print(charges_clean[500:510])
print(age_sorted[550:600])
```

Lists containing numbers as strings are replaced with lists containing integers/floats.

## Analysis

```python
# calculate average for some number data
def average(data):
    sum_of_data = 0
    for d in data:
        sum_of_data += d
    average = round(sum_of_data / len(data), 2)
    return average
```

## Age

```python
# average age 
print("The average age is: " + str(average(age_clean)) + " years.")
print()

# new list for age to charge comparison
age_to_charge = list(zip(age_clean, charges_clean))
print(age_to_charge[500:510])

# function for age to charge comparison
def avg_cost_for_age(data):
    young_cost = []
    young_n = 0
    adult_cost = []
    adult_n = 0
    senior_cost = []
    senior_n = 0
    for age, cost in data:
        if age > 0 and age < 30:
            young_cost.append(cost)
            young_n += 1
        elif age >= 30 and age <= 60:
            adult_cost.append(cost)
            adult_n += 1
        else:
            senior_cost.append(cost)
            senior_n += 1
    cost_young = average(young_cost)
    cost_adult = average(adult_cost)
    cost_senior = average(senior_cost)
    
    summary_young = print("There are " + str(young_n) + " people younger than 30 years of age. The average insurance cost for them is: " + str(cost_young) + " dollars.")
    summary_adult = print("There are " + str(adult_n) + " people between 30 and 60 years of age. The average insurance cost for them is: " + str(cost_adult) + " dollars.")
    summary_senior = print("There are " + str(senior_n) + " people older than 60 years of age. The average insurance cost for them is: " + str(cost_senior) + " dollars.")

avg_cost_for_age(age_to_charge)
print()

# new list for age categorization
age_to_gender = list(zip(age_clean, sex))
print(age_to_gender[:5])

# function for age categorization
def categorize_gender_age(data):
    young_male = 0
    young_female = 0
    adult_male = 0
    adult_female = 0
    senior_male = 0
    senior_female = 0
    for age, sex in data:
        if sex == 'male' and age < 30:
            young_male += 1
        elif sex == 'female' and age < 30: 
            young_female += 1
        elif sex == 'male' and age >= 30 and age <= 60:
            adult_male += 1
        elif sex == 'female' and age >= 30 and age <= 60:
             adult_female += 1
        elif sex == 'male' and age > 60:
            senior_male += 1
        else:
            senior_female += 1
    statement = print("There are " + str(young_male) + " males younger than 30 years of age." + "\n" \
                      "There are " + str(young_female) + " females younger than 30 years of age." + "\n" \
                      "There are " + str(adult_male) + " males between 30 and 60 years of age." + "\n" \
                      "There are " + str(adult_female) + " females between 30 and 60 years of age." + "\n" \
                      "There are " + str(senior_male) + " males older than 60 years of age." + "\n" \
                      "There are " + str(senior_female) + " females older than 60 years of age.")
    return statement

categorize_gender_age(age_to_gender)  
print()

# new list to determine average parent age
potential_parents = list(zip(children_clean, age_clean))
print(potential_parents[:5])

# function to determine average parent age
def average_parent_age(data): 
    total_age = 0
    parents = 0 
    for child, age in data:
        if child >= 1:
            total_age += age
            parents += 1
        else:
            pass
    avg_parent_age = round((total_age / parents), 2)
    statement = print("The average age of a person having at least one child is: " + str(avg_parent_age) + " years.")

average_parent_age(potential_parents)
```

## Sex

```python
# function that finds the male to female ratio
def gender_ratio(data):
    males = 0
    females = 0
    for d in data:
        if d == 'male':
            males += 1
        else:
            females += 1
    ratio = print("The male to female ration is: " + str(round(males / females, 2)))
    return ratio

gender_ratio(sex)
print()

# function that counts how many men and women are in the dataset
def count_genders(data):
    males = 0
    females = 0
    for d in data:
        if d == 'male':
            males += 1
        else:
            females += 1
    statement = print("There are " + str(males) + " males and " + str(females) + " females in our dataset.")
    return statement

count_genders(sex)
print() 

# counting men and women simplified
males = sex.count('male')
females = sex.count('female')
print(males, females)
```

## BMI

```python
# average BMI 
print("The average BMI index is: " + str(average(bmi_clean)))
print()

# new list to determine BMI per gender
gender_to_bmi = list(zip(sex, bmi_clean))
print(gender_to_bmi[:5])

# function to determine BMI per gender
def avg_bmi_per_gender(data):
    female_bmi = []
    male_bmi = []
    for gender, bmi in gender_to_bmi:
        if gender == 'male':
            male_bmi.append(bmi)
        else:
            female_bmi.append(bmi)
            
    avg_female = average(female_bmi) 
    avg_male = average(male_bmi)
    statement = print("The average BMI value for females in this dataset is: " + str(avg_female) + ".\n" \
                      "The average BMI value for males in this dataset is: " + str(avg_male) + ".")
    return statement

avg_bmi_per_gender(gender_to_bmi)   
print()

# BMI categorization ; taken from google
#    BMI	     Weight Status 
#Below 18.5	     Underweight
#18.5—24.9	     Healthy Weight
#25.0—29.9	     Overweight
#30.0 and Above  Obesity

# new list to categorize BMI and corresponding average age
age_bmi = list(zip(age_clean, bmi_clean))

# function to categorize BMI and corresponding average age
def bmi_sort(data):
    underweight = 0
    underweight_avg_age = []
    healthy_weight = 0
    healthy_avg_age = []
    overweight = 0
    overweight_avg_age = []
    above_obesity = 0
    above_obesity_avg_age = []
    for a, b in data:
        if b < 18.5:
            underweight += 1
            underweight_avg_age.append(a)
        elif b >= 18.5 and b <= 24.9:
            healthy_weight += 1
            healthy_avg_age.append(a)
        elif b >= 25.0 and b <= 29.9:
            overweight += 1
            overweight_avg_age.append(a)
        else:
            above_obesity += 1
            above_obesity_avg_age.append(a)
            
    avg_age_underweight = average(underweight_avg_age)
    avg_age_healthy = average(healthy_avg_age)
    avg_age_overweight = average(overweight_avg_age)
    avg_age_above_obesity = average(above_obesity_avg_age)
    
    statement = print("Underweight: " + str(underweight) + " - Average age: " + str(avg_age_underweight) + " years." + "\n" \
                      "Healthy weight: " + str(healthy_weight) +" - Average age: " + str(avg_age_healthy) + " years." + "\n" \
                      "Overweight: " + str(overweight) + " - Average age: " + str(avg_age_overweight) + " years." + "\n" \
                      "Above obesity: " + str(above_obesity) + " - Average age: " + str(avg_age_above_obesity) + " years.")
    return statement

bmi_sort(age_bmi)
```
